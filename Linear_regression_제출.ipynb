```
import numpy,pandas as pd
import matplotlib, matplotlib.pyplot as plt
import sklearn.linear_model as lm
health = pd.read_csv('/content/drive/My Drive/hufs/Data Mining/최종 데이터/Demacia 3차 linear 최종 수정.csv').set_index("plas") #데이터를 읽음
health.index = pd.to_numeric(health.index)# string 을 int 형으로 전환
health_linear = health.loc[health.index >  pd.to_numeric('-1')]

olm = lm.LinearRegression() 
x = numpy.array([x for x in health_linear.index])[:,numpy.newaxis]
y = numpy.array([y for y in health_linear['pres']])[: numpy.newaxis]
olm.fit(x,y)

numpy.shape(x)
yp = olm.predict(x)
olm_score = olm.score(x,y)
print(olm_score)

matplotlib.style.use("ggplot")
plt.scatter(health_linear.index, y)
plt.plot(health_linear.index,yp,color = 'blue')

plt.title("Linear Regression")
plt.xlabel("plas")
plt.ylabel("pres")
```
